# Section 1. Identification issues in macroeconomics

The main ideas are based on the following paper: **Emi Nakamura and JÃ³n Steinsson, 2018. "Identification in Macroeconomics,"** Journal of Economic Perspectives, American Economic Association, vol. 32(3), pages 59-86, Summer.

**1.1 Motivation**

Many macroeconomic questions have not changed a lot since macroeconomics was separated into a sub-discipline of economics; macroeconomists are still concerned with the sources of business cycle fluctuations, the effects of fiscal and monetary policies, the causes for unemployment and inflation growth, etc. Nakamura and Steinsson think that the reason of continuous relevance of these questions is that identification process in macroeconomics is challenging. Even though the last decade was called "revolution of identification" in the applied fields of economics (Angrist and Pischke, 2010b), the new research designs for causal effects estimation are still not completely applicable in macroeconomics and there are the reasons for that. Macroeconomic identification issues have the number of distinctive characteristics:

- *the variation in macroeconomic policies is rarely exogenous,* for instance, monetary policy is endogenized as possible: it is designed to react to the developments affecting output and inflation the most. The difficulty in identification can be seen on the example when the Federal Reserve lowers interest rates while an economy is heading to a downturn. If a researcher decides to run a simple regression of the change in output on the change in interest rates he is mostly likely to obtain the result that reduction in interest rates decreases the output. Of course, it will not be correct to rely on this result since the assignment to treatment is not exogenous in this case. The same works for estimating the effects of macroeconomic shocks; it can be seen on the example of credit intensity and severity of financial recessions which I will discuss in the second section. The authors of the paper claim that countries with higher credit intensity tend to have more severe financial recessions. However, causal interpretation here is difficult: credit intensity in the economies is not randomly assigned but it is caused by a large number of factors. Therefore, endogeneity of macroeconomic events is one the barriers to causal effects identification;

- *truly randomized natural experiments are normally not available.* For the reasons discussed above it is hard for the researchers to find the cases when assignment to the macroeconomic policies or shocks was random. However, exceptions also exist and I will describe the results of one of the natural experiment paper in the fourth section;

- *macroeconomic policies and shocks have a dynamic structure.* That means that the decisions taken in one period define the conditions for the future states. Moreover, dynamic nature makes the policies highly dimensional: some instruments of the policies affect the variables in a short run, some - in both short and long run; and some - only in few years in the future. The degree of persistance has a great impact on the future output and inflation;

- *the effects of fiscal, tax and monetary policies are mutually dependent*. For example, the effects of fiscal shocks depend on the response of monetary policy (whether monetary policy is constrained by the zero-lower-cound or not) and tax policy ( whether a war is financed with an increase in tax or with a debt which is planned to be paid off few years later);

- *the effects of macroeconomic policies and shocks depends on various characteristics of an economy*, for instance, what the level of slack in the economy is or how open the economy is. This problem can be also interepreted as agents' heterogeneity problem in microeconomics. However, the additional challenge in macroeconomics is that the number of characteristics the economies differ might be huge and it is simply impossible to control for all of them because of the data availability issues. We will see the example of this case in the second section;

- *expectations define the effects of the policies*. The extent to which a policy action is a surprise to households, investors or businesses affects how strongly the economy reacts to it. Moreover, it is not easy to catch the expectations in the data and the researchers are forced to use the results of the surveys which reliability is questionable.

**1.2 Solutions**

Due to the challenges described above causal effects identification in macroeconomic is more structural by nature. The researchers build the theoretical models to follow the shocks and predict their effects on the economy. However, the empirical methods for causal inference we have discussed during Microeconometrics course are not useless in macroeconomics; on the contrary, their relevance has grown in the last decade. In this subsection I will describe some of the approaches the researchers have followed recently to identify the causal effects in macroeconomics and then I will explain the content of the subsequent sections.

The empirical approaches to the identification in macroeconomics can be distinguished into two groups: aggregate and cross-sectional identification. In case of the *aggregate approach* the researchers estimate the effect of a policy or a shock on the country level and the estimates result can be used directly in taking the decisions about future policies. One example of this approach is measuring the government spending multiplier considering the evidence from wars by Barro and Redlick (2011). The main assumption of Barro and Redlick is that war-related defense spending is exogenous to country's output: WWI, WWII and Korean War have not been started because the US economy was in boom or recession at that time. Therefore, the authors regress the change in output on the change in defense spending in the periods of major wars. However, to realize this strategy the wars should affect output only through spending what is unlikely to be the case. Barro and Redlick also discuss this issue and provide two main possible confounding factors: patriotism and war-time controls as rationing and time controls. Since such a factor as patriotism is difficult to measure, that is why the resulting estimate might be biased. Another example of the aggregate identification approach is the paper of Blanchard and Perotti (1999). They regress government spending on four quartely lags of itself, taxes, output and a quadratic time trend. Next, they assume that the residual of this regression is exogenous to government spending and use it further to construct an impulse response function by iterating forward a VAR. The main weak spot here is again the assumption that controling for four quartely lags of itself, taxes, output and a quadratic time trend eliminates all the endogeneity. This weak spot tends to be one of the main barriers for the researchers to estimate the treatment effect using the aggregate approach. I am going to provide the example of this approach in the second section using the paper of Jorda et al. (2013).

The second approach is *cross-sectional identification.* Following this approach the researchers use geographically disaggregated data sets, for instance, on area, state or even country level, to estimate the regional causal effects. This approach allows to use a wide range of methods such as difference-in-difference and instrumental variable estimation. The example is Mian and Sufi (2014) paper where the authors compare the changes in employment in the cities which experienced large house price change with the cities which experienced smaller house price changes. The main challenge here is that, as the change in the house prices can cause change in employment in this area through the decomposition of spendings channel, as the employment can change the house prices in the area: reverse causality comes into play. To solve the issue Mian and Sufi use as an instrument housing supply elasticities estimates constructed by Saiz (2010). They find that the elasticity of employment with respect to housing net worth is 0.37. However, this estimate does not directly answer the question about the aggregate effect of house prices on employment because it was estimated comparing one city to another. That is the common case using cross-sectional identification: the estimates are not directly interpretable. However, it does not mean the results are not valuable; on the opposite, the researchers use them further to apply in theoretical frameworks and distinguish between different classes of models. I show the realization of this principle in the third section of my project based of the Nakamura and Steinsson (2014) paper.

**2.4 Causal graph and empirical strategy**

The identification strategy can be clearly represented with the help of the causal graph:

<img src="causal graphs/graph1.PNG">

Without a recession we can assume that excess credit has a positive effect on output following the explanation of Tobin (1989). However, if a normal or financial recession happens credit becomes an amplifier of a recession since it represents the fragility of financial system. Therefore, additionally credit affects output through normal or financial recession if the one happens. If everything was that simple as it was described then it could be possible to control for both recessions (or, by another words, block two back-door paths) and find the clear effect of credit on output in subsequent years when normal and financial recessions happen. However, credit intensity of an economy is not randomly assigned: the whole bunch of factors determine how leveraged an economy in some partcular country and year is. These factors are shown on the graph as macrocontrols and they affect all the variables in this environment. If it is the case that all of them are observable and available in the data then it is possible to block all the back-door paths and obtain the causal effect of credit on output after the recessions. The authors are using few macrocontrols and their lags for this purpose and I will discuss further whether it is possible to give a causal interpretation of the effect in this situation.

The authors use 5 empirical specifications through the paper, I am going to focus on the main one:

$$ \Delta_h Y_{it}  = \alpha + \gamma_i + \beta_1 pk(n)_{i,t} + \beta_2 pk(f)_{i,t} + \beta_3 pk(n)_{i,t}*excredit_{i,t} + \beta_4 pk(f)_{i,t}*excredit_{i,t} + controls_{i,t} + \epsilon_{i,t}, $$ 

where:

- $\Delta_h Y_{it}$ - change in log real GDP per capita of country i in year h relative to the year t;
- $\gamma_i$ - country fixed effects;
- $pk(n)_{i,t}, pk(f)_{i,t}$ - dummy variables which are equal to 1 if normal or financial recession happens in country i in year t; 0 otherwise;
- $excredit_{i,t}$ - dummy variable which is equal to 1 if there is higher than average credit intensity in economy i in year t;
- $controls_{i,t}$ - macrocontrols which include growth rate of real GDP per capita, CPI inflation rate, short-term interest rates on government securities (usually 3 months or less in maturity), long-term interest rates on government securities (usually 5 years or more in maturity), growth rate in investment to GDP ratio and their lags.

As the example I take the situation when there is a normal recession in country i in year t ($pk(n)_{i,t} = 1$) and treatment variable of interest here is excess credit ($excredit_{i,t}$).

Following the theoretical framework of Angrist and Pischke (2010a) conditional expectation with different treatment status:

$$E[\Delta_h Y_{it}|excredit_{i,t}=1] = \alpha + E[\gamma_i|excredit_{i,t}=1] + \beta_1 + \beta_3 + E[controls_{i,t}|excredit_{i,t}=1] + E[\epsilon_{i,t}|excredit_{i,t}=1]$$

$$E[\Delta_h Y_{it}|excredit_{i,t}=0] = \alpha + E[\gamma_i|excredit_{i,t}=0] + \beta_1 + E[controls_{i,t}|excredit_{i,t}=0] + E[\epsilon_{i,t}|excredit_{i,t}=0]$$

so then 

$$E[\Delta_h Y_{it}|excredit_{i,t}=1] - E[\Delta_h Y_{it}|excredit_{i,t}=0] =\beta_3 + E[\epsilon_{i,t}|excredit_{i,t}=1] - E[\epsilon_{i,t}|excredit_{i,t}=0]$$



Then $\beta_3$ is a treatment effect of excess credit on output after the normal recession and $E[\epsilon_{i,t}|excredit_{i,t}=1] - E[\epsilon_{i,t}|excredit_{i,t}=0]$ is a potential selection bias. It disappears only under the condition that $E[\epsilon_{i,t}|excredit_{i,t}=1] = E[\epsilon_{i,t}|excredit_{i,t}=0]$; equivalently, the error term is independent from the assignment to treatment. It means there are no more characteristics in the error term that define the credit intensity of an economy.

Therefore, *coefficient $\beta_3$ can be given causal interpretation only under the condition that all the variables representing the characteristics of economies and potentially affecting credit intensity are included into the controlling terms.* 

In principle, there can be hundreds of political, economic, financial and social factors that define how leveraged an economy is. In my regression I am using the following financial controls: growth rate of real GDP per capita, CPI inflation rate, short-term interest rates on government securities (usually 3 months or less in maturity), long-term interest rates on government securities (usually 5 years or more in maturity), growth rate in investment to GDP ratio and their lags. Obviously, it is hard to assume that they fully describe all the conditions determining the level of leverage in the economy. Moreover, if a researcher attempts to gather a data on all possible conditions it seems to be a challenging task: first of all, the huge amount of data is necessary. Secondly, some of conditions are hard to measure, for instance, attitude of the population to the loans and leverage. That becomes one of barriers for macroeconomists in estimating causal effects empirically on the aggregate level and that is why they turn to another methods such as cross-sectional identification.

# Section 3. Estimating regional multiplier for government spending modelling

The following section replicates the main results of the paper: **Emi Nakamura and JÃ³n Steinsson, 2014. "Fiscal Stimulus in a Monetary Union: Evidence from US Regions."** American Economic Review, 104 (3): 753-92.

**3.1 Introduction**

Government spending include the spending on the health care, national defense, social security, unemployment disabilities, education, etc. In this paper the authors focus their attention on evaluation the economic effect of miltary spending and in this respect military spending is a good illustration of a policy the effect of which is difficult to measure. The reason is that military spendings are rarely exogenous: they result from the interplay of a bunch of political and economic factors (Mintz, 1992). Mintz provides three main approaches in the studies of military spending based on the US history; the first one is so-called "the arms race" approach: the country reacts to the behavior of rival countries and starts to buildup military power as a response to a possible threat. This line can be followed in the USâUSSR arms race in the post-World War II era. The second approach presents the idea that military budgetary policy relies heavily on the past expenditures with only marginal adjustments of previous appropriations. This approach has a name of incremental policies. The third approach integrates the fields of national security and political economy. In this case the military spending is determined by such factors as an American role in the global political and economic system, political and electoral cycles, international security issues, etc. Therefore, it is hard to call military spending an exogenous policy, that is why the scholars use various ways to estimate its real effect on an economy (the examples of the surveys are Perotti (2007), Hall (2009), Cogan et al. (2010)).  
   
   The effect of government spending is normally measured by a multiplier: the percent increase in output when government spending is increased by 1 percent GDP. In their paper Nakamura and Steinsson make an attempt to measure military spending multiplier but on state level: the effect of an increase in military spending in a region relative to the aggregate level on relative output and employment. They call this indicator an "open economy relative multiplier", opposite to the "closed economy relative multiplier" which is measured simply on the aggregate level. Afterwards, they use estimated result for an open economy relative multiplier in theoretical framework to assess how it is related to a multiplier on the the aggregate level. Such an exersise helps the authors to distiniguish between different macroeconomic models and to make a conclusion about which models tend to do a better job in explaining macroeconomic environment. In this section I am going to replicate the main results of the first part on estimating an open economy relative multiplier and then to review authors' main findings on macroeconomic modelling.

**3.2 Empirical strategy**

*Main empirical specification and issues for identification*

The main empirical specification estimating the percent increase in output on the state level when government spending is increased by 1 percent GDP on the state level:

$$\frac{Y_{i,t} - Y_{i,t-1}}{Y_{i,t-1}} = \theta + \alpha_i + \gamma_t + \beta \frac{G_{i,t} - G_{i,t-1}}{Y_{i,t-1}} + \epsilon_{i,t},$$ 
where:
- $Y_{i,t}$ - per capita output in state i in year t
- $G_{i,t}$ - per capita military procurement spending in state i in year t
- $\alpha_i$ - state fixed effects which allow for state specific trends in military procurement
- $\gamma_t$ - time fixed effects which allows for macro shocks and policies (changes in taxation and monetary policy)
- $\theta$ - constant
- the authors also cluster standards errors by state to account for overlapping nature of observations.

However, as it was mentioned earlier military spending is not an exogenous policy. The same works for state level too: state military spending is definied by various political and economic factors. One example of endogeneity can be that those states which do worse economically need an investment the most and the officials find it easier to negotiate for more spending. In this case treatment effect measure will suffer from selection bias. This can be described using the theoretical framework of Angrist and Pischke (2010a):  

Slight transformation of specification as an example:

$$\Delta Y_{i,t} = \theta + \alpha_i + \gamma_t + \beta \Delta G_{i,t} + \epsilon_{i,t},(1) $$     

Let treatment variable $\Delta G_{i,t}$ = 1 if government spending is increased by 1 percent GDP on the state level in period t; 0 otherwise (I make treatment variable discrete only as a matter of an example. Futher in estimation treatment variable is continuos). Evaluating conditional expectation with different treatment status gives:

$$E[\Delta Y_{i,t}|\Delta G_{i,t} = 1] = \theta + E[\alpha_{i}] + E[\gamma_{t}] + \beta + E[\epsilon_{i,t}|\Delta G_{i,t} = 1] $$
$$E[\Delta Y_{i,t}|\Delta G_{i,t} = 0] = \theta + E[\alpha_{i}] + E[\gamma_{t}] + E[\epsilon_{i,t}|\Delta G_{i,t} = 0] $$

so that $$E[\Delta Y_{i,t}|\Delta G_{i,t} = 1] - E[\Delta Y_{i,t}|\Delta G_{i,t} = 0] = \beta + E[\epsilon_{i,t}|\Delta G_{i,t} = 1] - E[\epsilon_{i,t}|\Delta G_{i,t} = 0], $$

where $\beta$ is treatment effect and $E[\epsilon_{i,t}|\Delta G_{i,t} = 1] - E[\epsilon_{i,t}|\Delta G_{i,t} = 0]$ is selection bias. Thus, selection bias takes place because of the correlation between error term and treatment variable: military spending is not randomly assigned. Moreover, since economically poorly doing states might be more likely to obtain bigger piece of aggregate spending on military purposes one can expect a negative selection bias in treatment effect on state output.  As we will see later running the simple OLS regression that is exactly the case.

*IV Identification strategy*

Nakamura and Steinsson propose the following solution to the identification problem. Figure 3.1 presents the evolution of state military procurement spending relative to state output for California and Illinois as well as aggregate spending relative to aggregate output.

```python
fig31()
```

Firstly, it can be noted that variation in military spending is driven by important geopolitical events: Vietnam War, Soviet Envasion in Afganistan, 9/11. Secondly, military spending in California is systematically sensitive to the movements of national spending than military spending in Illinois. The authors use this **variation in sensivity as an instrument** to identify the effects of state military spending in IV regression. **Main identifying assumption**: the US does not start a military biuldup on the national level because some of the states are doing poorly relative to the other states. I can think of it on the example of Vietnam war: the military budget on the aggregate level is based on threat evaluation or other military/political factors but doubtably  on the economic conditions of some particular states relative to the others.

So then the first stage empirical specification can be presented as the following:

$$\frac{G_{i,t} - G_{i,t-1}}{Y_{i,t-1}} = \theta + \alpha_i + \eta_i(state_i*\frac{G_{t} - G_{t-1}}{Y_{t-1}}) + \mu \frac{G_{t} - G_{t-1}}{Y_{t-1}} + \epsilon_{i,t},$$
where:
- $G_{i,t}$ - per capita military procurement spending in state i in year t
- $G_{t}$ - aggregate per capita military procurement spending in year t
- $Y_{i,t}$ - per capita output in state i in year t
- $Y_{t}$ - aggregate per capita output  in year t
- $\alpha_i$ - state fixed effects which allow for state specific trends in military procurement
- $\eta_i$ - effects allowing for different sensivities accross different states
- $\theta$ - constant

Slight transformation of the first stage specification:
$$\Delta G_{i,t} = \theta + \alpha_i + \eta_i(state_i*\Delta G_{t}) + \mu \Delta G_{t} + \epsilon_{i,t}, (2)$$

It estimates the sensivity of military spending across the states as a response to national military buildup.

*Wald estimator*

Now I would like to show how the instrument comes into play and helps to estimate the real treatment effect. Using both theoretical frameworks of Angrist and Pischke (2010a) and Morgan and Winship (2007) I construct the Wald formula which shows a transparent implementation of IV strategy. 

$state_i*\Delta G_{t}$ or the change in national spending as a fitted values for each state plays a role of the instrument in the paper. As a matter of example I write it as $Z_{i,t}$. Let $Z_{i,t}$ = 1 if national government spending is increased by 1 percent GDP in the period t; 0 otherwise.

Conditional expectation of second stage regression in the equation (1):

$$ E[\Delta Y_{i,t}] = \theta + E[\alpha_i] + E[\gamma_t] + \beta  E[\Delta G_{i,t}] + E[\epsilon_{i,t}] $$

Writing it as a difference equation with an instrument $Z_{i,t}$:

$$ E[\Delta Y_{i,t}|Z_{i,t} = 1] - E[\Delta Y_{i,t}|Z_{i,t} = 0]= \beta (E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]) + (E[\epsilon_{i,t}|Z_{i,t} = 1] - E[\epsilon_{i,t}|Z_{i,t} = 0])$$

Dividing both sides by $E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]$ yields:

$$\frac{E[\Delta Y_{i,t}|Z_{i,t} = 1] - E[\Delta Y_{i,t}|Z_{i,t} = 0]}{E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]} = \frac{\beta (E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]) + (E[\epsilon_{i,t}|Z_{i,t} = 1] - E[\epsilon_{i,t}|Z_{i,t} = 0])}{E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]}$$

Then Wald estimator is equal to: 

$$\frac{E[\Delta Y_{i,t}|Z_{i,t} = 1] - E[\Delta Y_{i,t}|Z_{i,t} = 0]}{E[\Delta G_{i,t}|Z_{i,t} = 1] - E[\Delta G_{i,t}|Z_{i,t} = 0]} = \beta_{IV, WALD}$$

under one condition: $E[\epsilon_{i,t}|Z_{i,t} = 1] = E[\epsilon_{i,t}|Z_{i,t} = 0]$ or, equivalently, error term $\epsilon_{i,t}$ in the main equation (1) is uncorrelated with an instrument $Z_{i,t}$. The main assumption of the authors that the US does not embark military biuldup on the national level based on economic conditions in some states relative to the other states justifies this condition. So then exogeneity conditions for the instrument  $E[\epsilon_{i,t}|Z_{i,t}] = 0$ and $E[\Delta G_{i,t}|Z_{i,t}] \neq 0$ are satisfied.

In theory, it is possible to block the back-door path G $\dashleftarrow \epsilon \dashrightarrow$ Y since the path a is fork of mutual dependence. However, state military spending and state output might be definied by the huge number of factors and not all of them are observable to a researcher. For instance, cultural attitudes towards war and also economic development can motivate the officials to argue for more or less military spending in the state. That is why Nakamura and Steinsson propose to use instrumental variable Z - variation in sensitivity of state military spending to the movements in national spending. They assume that national military spending is independent from the state output what guarantees the absense of the path Z $\rightarrow$ Y. Therefore, using Z they identify the clear effect of G on Y.

**3.6 Implementation in macroeconomic models**

To interpret open economy multiplier and relate it to the multiplier on aggregate level the authors develop a theoretical baseline framework, calibrate the fundamentals, describe the classification of the specification models, simulate the data from the baseline models, run the regressions to derive open and closed economy multipliers and, finally, discuss the implications of macroeconomic models in predicting government spending effect.

The first step is to build a framework which allows for further adjustment to different classes of macroeconomic models. Since in the paper states play a role of open economies in such a monetary union as the US the authors rely heavily on international economics models. There are two types of regions: home where the shock accours and foreign as the rest of the economy. The model includes three types of agents. Households supply labor on the market, consume home and/or foreign good and have access to financial markets. Each period they decide how much labor to supply, how much to consume, how much of each differentiated good to consume and what portfolio of assets to buy. Government relies on lump-sum taxes, spend and purchase, conducts monetary and fiscal policies. Firms maximize their value specializing on the production of differented good and employing labor as the only production factor. The authors use a bunch of sources and datasets to calibrate the fundamentals - preferences and technologies.

Employing above described baseline model Nakamura and Steinsson consider several specifications for the fundumentals (separable or GHH preferences, flexible or sticky prices) as well as specifications for monitary and fiscal policies. In the Neoclassical model with flexible prices money is neutral and, therefore, specification for monetary policy is irrelevant. Tax policy, however, is important. In the New Keynesian model with sticky prices the situation is the opposite: monetary policy plays a role rather than fiscal policy. The monitary policies include Volcker-Greenspan policy where an authority aggressively raises real interest rate to decrease inflation effects of government spending shock; fixed real-rate policy where government maintains a fixed real rate in response to government spending shock; and fixed nomial-rate policy where government maintains a fixed nominal rate in response to government spending shock. In the last two kinds of policies government responds aggressively to the inflation effects of all other shocks, however. There are also two types of tax policies. The first one is constant income tax rates policy in which government spending is financed by lump-sum taxes. The second one is balanced budget policy in which increase in government spending is associated with distotionary taxation.

After distinguishing between 5 classes of models the authors turn to multipliers analysis. To obtain closed and open economy multipliers they simulate the data from the baseline model described above in Matlab and run the regressions on the equation (1) for state and aggregate levels and for different classes of models.  The results for the Neoclassical and the New Keynesian models with separable preferences can be seen in the following table:

The authors describe each obtained multiplier and the reasons why it happens to be higher or lower. I am not going to do that in details since it is a matter of macroeconomic modelling and, hence, far beyond the frameworks of my project.  I will rather concentrate on the core results. 

The separable preferences mean that consumption and labor enter utility specifications separatably and they are neither complements nor substitutes. In case of such preferences assumption the open economy multiplier for two types of fiscal policy in the Neoclassical model is equal to 0.43. That means that the multiplier is almost insensitive to the aggregate tax policy. In case of New Keynesian model sensivity is higher - 0.83. Why does the sensivity differ so much in two models? The government spending shock at zero lower bound forces prices to raise. Since inthe  Neoclassical model the prices are flexible they jump immediately as a response to a shock and then start to fall down. That forces interest rate to raise and private spending to slow down. On opposite, in the New Keynesian model prices are rigid in a ashort run, interest rate falls as a response to a shock and that boosts private spending. That explains the difference in the multiplier across the models. However, the obtained result in two models is still too low for out estimated multiplier of 1.5. GHH preferences mean that labor and consumption can suplement each other. It implies an extra consumption, for instance, food away from home, gas at work, etc. It will also have a powerful implication of spending mulplier since the households will have to work more to produce an additional output. That will raise consumption and output. Looking at the results of the regression the Neoclassical open economy multiplier is still insensitive to the fiscal policy, the New Keynesian model one, however, is doing a good job in predicting estimated open economy multiplier - 1.42. Intuitively, that happens because due to complementarity between labor and consumption households want to shift their consumption to the periods of high work what results in positive government spending shocks. Therefore, we can make a conclusion that the New Keynesian model with GHH preferences reflects the estimated multiplier the most. 

What is about the main point of interest - closed economy multiplier - in this case? Under the circumstances of fixed and nominal rate policies the model can generate extremely high value for the multiplier. But if monetary policy is highly responsive to output the New Keynesian model produce similar result with the Neoclassical one. The main conclusion is that based on more precise multiplier prediction by the New Keynesian model demand shocks can have a large effect on the output if the government tend to fix the interest rate in response to a shock.

**3.7 Conclusion**

The effect of government spending on the output is difficult to measure because the policy is rarely exogenous. The authors of the paper are using the following approach to identify the effect of military spending. Firstly, they estimate so-called open economy multiplier what reflects the percent increase in output on the state level when government spending is increased by 1 percent GDP on the state level. But state military spending might be also endogenous. One example is that those states which do worse economically need an investment the most and the officials could find it easier to negotiate for more spending. To overcome the selection bias problem the authors employ instrumental variable strategy. The main assumption here is that aggregate military spending is exogenous to the factors defining output of the states. The authors also state that the sensitivity of the state spending to the movements of national spending differs across the states but it is persistant across the time. They derive this variation in the first stage regression and use it in the second stage regression to estimate the effect of state military spending including state and time fixed effects and clustering the standard errors. The obtained regional multiplier is equal to 1.4.

However, the open economy multiplier does not directly answer the question about the effect of aggregate military spending on country's GDP. That is why Nakamura and Steinsson develop a theoretical framework to interpret the estimated regional multiplier and to make an inference about the aggregate multiplier. They build a baseline model and provide few specifications reflecting the different implementation of monitary and fiscal policies. Afterwards, using the simulated data they measure both open and closed economy estimators. That provides a possibility to compare the open economy multipliers estimated by real and simulated data and to conclude what model tends to make better predictions. It also allows to analyze the behavior of closed economy multiplier under different macroeconomic circumstances. Hence, the paper of Nakamura and Steinsson is a very clear example how the technics for identifying causal effects could be implemented on the macroeconomic level.

**4.4 Empirical strategy**

Main estimating equation:

$$C_{i_t+1} - C_{i,t} = \gamma_t + \beta ESP_{i,t+1} + \mu X_{i,t+1} + u_{i,t+1} (3),$$

where:
- $C_{i_t+1} - C_{i,t}$ - change in consumption of household i between periods t and t+1
- $\gamma_t$ - seasonal time fixed effects
- $X_{i,t+1}$ - control variables (age, change in the number of children, change in family size) to absorb preference-driven differences in consumption accross households

Variable $ESP_{i,t+1}$ takes the following forms dependable on specification:
- $ESP_{i,t+1}$ - total dollar amount of payment receieved by household i in period t+1
- $I(ESP_{i,t+1}) $ - dummy variable indicating whether any payment was received in the period t+1

Therefore, coefficient $\beta$ has an interpretation as the average effect of household  expenditure on stimulus payment corrected for households heterogeneity and seasonal variation in consumption. The authors also cluster standard errors to eliminate heteroskedasticity and within-household serial correlation (since the majority of households participate in the survey more than once).

Following the theoretical framework from the section 3.2 the treatment effect does not suffer from selection bias since in every period of the sample households are assigned to treatment randomly based on their two last SSN digits. I will not present the theoretical framework here because it is practically the same as in the section 3.2 with some changes in specification and variables names.

**Critical notes on HH characteristics:** 
- firstly, not all US population participated in the program of Economic Stimulus Payments but approximately 85% of "tax units". Some households did not satisfy basic eligibility criteria. Since the choice into those 85% was not random the interpretation of the obtained in the paper treatment effect as for the whole population will be a mistake. One should clearly understand interpreting the coefficient that it reflects a spending of only a particular part of population satisfying certain criteria.

- secondly, the coefficient for the treatment effect is estimated based on the CE survey sample. In the paper there is no information about how the households are selected into this sample and whether this selection is truly random. This issue needs more investigation and, once again, one should be careful with coefficient interpretation.

- thirdly, some percent of the households received ESP later than it was planned by the official schedule. It can be the case that the reasons of why some households received ESPs later can be correlated with households characteristics. The authors solve this issue in one of the regression droping this kind of observations from the sample.

**Critical notes on continuous effect:**

- the negative bias in the ESPs effect can appear because of the estimation strategy itself. The authors look at every period of the sample considering whether a household received ESP or not. In this case they estimate the change in consumer spending only in the the same period in which ESP was received. However, the ESP effect can last not only for the same period but for the future periods too. The example is when a household decides to spend ESP or its part in the next period what could be possible for a variety of reasons. The athouts deal with this issue estimating the effect on consumer spending in the period after receiving of ESP in one of their regressions. However, we cannot control for the effects in further periods what makes negative bias in the estimated ESP effect feasible.

From the section 3 we know that even though the obtained result might be not interpretable on the country level its value still can be used in theoretical modelling; the same works here. Although the findings are not related to any particular theoretical model, obtained result rejects the rational expectations life-cycle/permanent income hypothesis (LCPIH). According to Friedman (1957), the hypothesis says that individual's consumption does not depend on current income alone, but it also depends on the expectations about the future. Individuals consume based on their permanent income and small changes in the current income should not cause the large changes in the consumption. As we see from the results it is true since the households tend to spend 50-60% of ESP payments in the period of receipt. Besides, we can also test the Ricardian equivalence which states that stimulation of an economy by increasing government spending does not change the demand. It explained by that fact the households save this extra money to repay an expected tax increase in the future. Based on the results, the Ricardian equivalence can be also rejected.

# Section 5. Summary

The identification issues in macroeconomics have the number of distinctive characteristics: endogenous variation, rare possibilities to organize a natural experiment, complex dynamic structure, role of expectations and others. Following these challenges, in my project I ask myself a question: what are the empirical methods to estimate the causal effects in macroeconomics and how might the tools learned within [Microeconometrics course](https://github.com/HumanCapitalAnalysis) be employed to address the identification issues? I describe two empirical approaches which help macroeconomists to identify the effects. The first one, the aggregate approach, implies that the effects of a policy or a shock are measured on the country level and the results can be used directly in taking the decisions about future policies. The second approach, cross-sectional identification, implies that the researchers use geographically disaggregated data sets to estimate the regional causal effects and employ the results further in the theoretical modelling.

I provide three examples of how the macroeconomists deal with identification issues in their papers. The first example is the estimation of the effect of credit on severity of financial crises by Jorda et al.(2013). According to the results, the financial recessions tend to be more severe than the normal recessions. Moreover, the negative effect of a recession is amplified by excess credit in the economy: if the economy is highly leveraged the financial system is more fragile and exogenous shock has all the chances to "explode" in a systematic crisis. However, such a result is difficult to give a causal interpretation since the assignment to the treatment or, by another words, to the leverage level in the economy, is not exogenous; It is explained by a large number of factors and some of them are unobservable in the data. It is a common problem in macroeconomics, that is why the researchers turn to the cross-sectional identification.

As an example of the cross-sectional identification I use the paper of Nakamura and Steinsson (2014) aiming to estimate government spending multiplier. The authors employ the assumption that the US does not start a military biuldup on the national level because some of the states are doing poorly relative to the other states. This assumption allows them to refer to instrumental variable estimation and obtain regional government spending multiplier. They use the result further in building theoretical model with a possibility of making different assumptions. That gives them a tool to estimate open and closed economy multipliers with simulated data under New Keynesian and Neoclassical models, to imply the obtained open economy multiplier from the real data and to evaluate which model makes better a prediction. 

In the fourth section I show how a possibility of a natural experiment might be employed to estimate the effect of Economic stimulus payment in 2008 on households consumption. Here the structure of the tax cuts provides an amazing opportunity for natural experiment since that timing of the payments was randomly assigned by the recipient's SSN. The main result is that, on average, households spend 12-22% of their ESP on nondurable goods and 50-60% of ESP on all goods. This result can be also used further to confirm or reject some of macroeconomic theories. 

Obviously, I do not describe all the empirical methods that are used in macroeconomics to identify the causal effects. I show a a possible simple classification that allows to distinguish between different methods and to demonstrate their strong and weak spots.
